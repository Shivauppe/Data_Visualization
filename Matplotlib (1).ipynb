```python
x = np.linspace(0,10,100)
x
```

```python
plt.plot(x,np.sin(x),label = 'Sinewave',linewidth=2,linestyle = '--')
plt.plot(x,np.cos(x),label = 'Coswave',linewidth=2)
plt.legend()
plt.show()
```

```python
plt.plot(x,np.tan(x),'--',color = 'red')
```

```python
plt.plot(x,np.tan(x),'.-',color = 'gray')
```

```python
plt.plot(x,np.tan(x),':',color = 'pink',linewidth = 4)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,1)
plt.plot(x,np.cos(x))

#show 
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

#show 
plt.show()
```

```python
# 4plots

plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),label = 'sinewave',linestyle = '--',color = 'red')
plt.grid()
plt.legend()


plt.subplot(2,2,2)
plt.plot(x,np.cos(x),label = 'coswave',linestyle = '--',color = 'violet')
plt.grid()
plt.legend()


plt.subplot(2,2,3)
plt.plot(x,np.tan(x),label = 'tanwave',linestyle = '--',color = 'blue')
plt.grid()
plt.legend()


plt.subplot(2,2,4)
plt.plot(x,np.sin(x+1),label = 'Sine(x+1)wave',linestyle = '--',color = 'black')

plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,np.sin(x),label = 'Sine(x)wave',linestyle = '--',color = 'black')
plt.title("Sinewave plot")
plt.xlim(1,15)
plt.ylim(-1,1)
plt.axis("equal")
plt.legend()
plt.grid()
plt.show()
```

```python
#Pie chart >> to see the % distribution of a data

lang = ['C','Java','Python']
students = [20,40,100]

plt.pie(students,labels = lang,autopct = '%1.2f%%')
plt.title("Pie Chart")
plt.show()
```

```python
lang = ['C','Java','Python']
students = [20,40,100]
explode = (0.1,0.1,0.1)
plt.pie(students,labels = lang,autopct = '%1.2f%%',explode = explode)
plt.title("Pie Chart")
plt.show()
```

```python
lang = ['C','Java','Python']
students = [20,40,100]
explode = (0.1,0.1,0.1)
plt.pie(students,labels = lang,autopct = '%1.2f%%',explode = explode,shadow = True)
plt.title("Pie Chart")
plt.show()
```

```python
years = [2021,2018,2022,2023,2017]
category1 = [20,25,30,28,35]
category2 = [10,15,20,18,25]
category3 = [5,8,10,9,15]
```

```python
plt.stackplot(years,category1,category2,category3,labels = ["category1","category2","category3",],
              colors = ["skyblue","lightgreen","violet"])
plt.xlabel("Titiles")
plt.ylabel("years")
plt.title("Area plot")
plt.legend()
plt.show()
```
