```python
# seaborn numerical data sets
# relplot has options of plotting both categorical and numerical data
#relplot how the relationship between two variables 
# relplot >> relationship plot
```

```python
#how tip is varying wrt to total_bill >> scatterplot/by default relplot scatter plot
# Figure-level interface for drawing relational plots onto a FacetGrid.

sns.relplot(x = 'total_bill' , y = 'tip', data = df)
```

```python
sns.relplot(x = 'total_bill' , y = 'tip', data = df, hue = 'smoker') # relationship beetween two variables 
```

```python
sns.relplot(x = 'total_bill' , y = 'tip', data = df, hue = 'size') # relationship beetween two variables 
```

```python
# hue - diff color coding to give the more insights.
sns.relplot(x = 'total_bill' , y = 'tip', data = df, style = 'size') 
#style gives the diff reprentaion of diff columns
```

```python
sns.relplot(x = 'total_bill' , y = 'tip', data = df, style = 'smoker', hue = 'time') 
```

```python
#relplot for making a line plot >> generating a plot
#line plot >> to understand the trend in a data , a time column/sequence is generally is there
```

```python
import numpy as np
import pandas as pd
time = np.arange(500)
value = np.random.randn(500).cumsum()
df1 = pd.DataFrame({'time' :time ,'value ':value})
```

```python
df1
```

```python
sns.relplot(x =time, y=value, kind = 'line',data = df1, sort = True)
```

```python
sns.catplot(x = 'day', y = 'total_bill' , data = df)
# insights - Many people are visiting on weekends compared to weekdays
```
